# Python String Methods


```python
a = "Any simple string"

print("ORIGINAL STRING:", a)
print("-" * 40)
```

## 1️⃣ CHECKING / TESTING METHODS


```python
print("1. CHECKING METHODS")
print(a.isalpha())      # True if only alphabets (no spaces)
print(a.isdigit())      # True if only digits
print(a.isalnum())      # True if alphabets + digits only
print(a.isupper())      # True if all characters are uppercase
print(a.islower())      # True if all characters are lowercase
print(a.istitle())      # True if string is in Title Case
print(a.isspace())      # True if string contains only spaces
print("-" * 40)
```

## 2️⃣ CASE CONVERSION METHODS


```python
print("2. CASE CONVERSION METHODS")
b = a.lower()           # Converts to lowercase
c = a.upper()           # Converts to uppercase
d = a.title()           # Converts to title case

print(b)
print(c)
print(d)
print(b.islower())      # Check if b is lowercase


# capitalize() and swapcase()
print(a.capitalize())   # Capitalizes first character
print(a.swapcase())     # Swaps uppercase to lowercase and vice versa
print("-" * 40)
```

## 3️⃣ SEARCHING & COUNTING METHODS


```python
print("3. SEARCHING & COUNTING")
print(a.count('s'))     # Count occurrences of 's'
print(a.find('s'))      # First index of 's' (returns -1 if not found)
print(a.index('s'))     # First index of 's' (error if not found)
print("-" * 40)
```

## 4️⃣ STARTS WITH / ENDS WITH

```python
print("4. STARTS / ENDS CHECK")
print(a.startswith('A')) # Check starting character
print(a.endswith('g'))   # Check ending character
print("-" * 40)
```

## 5️⃣ STRIP METHODS (REMOVE FROM EDGES)


```python
print("5. STRIP METHODS")
print(a.strip("A"))     # Remove 'A' from both sides
print(a.lstrip("A"))    # Remove 'A' from left side
print(a.rstrip("A"))    # Remove 'A' from right side
print("-" * 40)
```

## 6️⃣ REPLACEMENT METHOD

```python
print("6. REPLACE METHOD")
print(a.replace('s', 'A'))  # Replace 's' with 'A'
print("-" * 40)
```


## 7️⃣ SPLIT & JOIN METHODS

```python
print("7. SPLIT & JOIN")

# split()
f = a.split(" ")         # Split string into list
print(type(f))
print(len(f))
print(f)

# partition()
e = a.partition(" ")     # Split into tuple of 3 parts
print(type(e))
print(e)

# join()
print(a.join(a))   # Join each character using 'a' as separator

print("-" * 40)
```

## 8️⃣ LENGTH & COMPARISON

```python
print("8. LENGTH & COMPARISON")
print(len(a))            # Length of string
print(b == c)            # Compare two strings
print("-" * 40)
```

## 9️⃣ ASCII / UNICODE CONVERSION

```python
print("9. ASCII / UNICODE")
print(ord('a'))          # Character to ASCII value
print(chr(98))           # ASCII value to character
print("-" * 40)
```

## Write a program to return sum of the digits present in string

```python
mystring = input("enter a string :")
sum=0
for num in mystring:
    if num.isdigit() :
        sum=sum+int(num)       
print(f"Sum of the digits is {sum}")
```

## Write a program to Count digits, alphabets, words, Special_Symbol in a string

```python
mystring = input("Enter a string : ")

digits = alphabets = spaces = special = 0

words = mystring.split()
print(f"No of words in the string : {len(words)}")

for ch in mystring:
    if ch.isdigit():
        digits += 1
    elif ch.isalpha():
        alphabets += 1
    elif ch.isspace():
        spaces += 1
    else:
        special += 1

print(f"Number of digits in the string : {digits}")
print(f"Number of alphabets in the string : {alphabets}")
print(f"Number of spaces in the string : {spaces}")
print(f"Number of special symbols in the string : {special}")
```

## Program to check if the String is Pallindrome or not

```python
mystring=input("Enter a string: ")

mystring = mystring.lower().replace(" ", "")

if mystring==mystring[::-1] :
    print("Entered string is a pallindrom")

else: 
    print("Entered string is not a pallindrom")
```

## Program to reverse the words in the String

```python
s = input("Enter a string: ")
# Split into words, reverse each word, then join them back with spaces
result = " ".join([word[::-1] for word in s.split()])
print(result)
```

## fINDING THE PRIME NUMBER 

```python
number = int(input("Enter the number: "))
divisors = []

for i in range(2, number):
    if number % i == 0:
        divisors.append(i)

if len(divisors) == 0:
    print(f"Given number {number} is a prime number")
else:
    print(f"Given number {number} is not a prime number")
    # Yahan join use karenge:
    # divisors list ke numbers ko string mein badal kar space se jodna
    div_str = " ".join([str(d) for d in divisors])
    print(f"Divisors are: {div_str}")
```
