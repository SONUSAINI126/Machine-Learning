```python
import numpy as np 
```

```python
# Logistic Regression

class LogisticRegression:
    def __init__(self,learning_rate=0.1,n_iter=1000):
        self.bias = None
        self.weights = None
        self.lr = learning_rate
        self.n_iter = n_iter
    def _sigmoid(self,z):
        return (1/(+np.exp(-z)))

    def fit(self,X,y):
        m,n = X.shape

        self.bias = 0
        self.weights = np.zeros(n)

        for i in range(self.n_iter):
            z = self.bias + np.dot(X, self.weights)
            y_pred = self._sigmoid(z)

            db = (1/m) * (y_pred-y)
            dw = (1/m) * np.dot(X.T , (y_pred-y ))

            self.bias -= self.lr * db
            self.weights -= self.lr * dw
            
    def get_probabilities(self,X):
        z = self.bias + np.dot(X, self.weights)
        return self._sigmoid(z)
        
    def predict(self,X,thresold=0.5):
        probabilities = self.get_probabilities(X)
        y_pred_bool = probabilities >= thresold
        return y_pred_bool.astype(int)
```

```python
X = np.array([[1,2] , [2,3] , [3,4] , [4,5]])
y = np.array([0,0,1,1])

model = LogisticRegression()
model.fit(X,y)

y_pred = model.predict(X)
print(y_pred)

print(model.get_probabilities(X))
```
